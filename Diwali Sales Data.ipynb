```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding= "unicode_escape")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"Shaadi"})
```

```python
df.describe()
```

# EDA

```python
df.columns
```

```python
a=sns.countplot(x="Gender",data=df)
plt.savefig("Gender.jpg")

for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)
plt.savefig("gender with amt.jpg")

plt.show()
```

```python
a=sns.countplot(data=df,x="Age Group",hue="Gender")
plt.savefig("age grp.jpg")

for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_sta=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(16,5)})
sns.barplot(x="State",y="Orders",data=sales_sta)
plt.savefig("state orders.jpg")
plt.show()
```

```python
sales_amt=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(16,5)})
sns.barplot(x="State",y="Amount",data=sales_amt)
plt.savefig("state amount.jpg")

plt.show()
```

```python
a=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(4,4)})
plt.savefig("maritalstatus amt1.jpg")
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_mar=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(4,4)})
plt.savefig("maritalstatus amt.jpg")

sns.barplot(x="Marital_Status",y="Amount",data=sales_mar,hue="Gender")
```

```python
sns.set(rc={"figure.figsize":(20,5)})
a=sns.countplot(data=df,x="Occupation")
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_sta=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(25,4)})
plt.savefig("Occupation amt.jpg")
sns.barplot(x="Occupation",y="Amount",data=sales_sta)
```

```python
a=sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(20,5)})
plt.savefig("pc.jpg")
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_pro=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
plt.savefig("Product category.jpg")
sns.barplot(x="Product_Category",y="Amount",data=sales_pro)
```

```python
sales_pro=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
plt.savefig("ProductID with orders.jpg")
sns.barplot(x="Product_ID",y="Orders",data=sales_pro)
plt.show()
```
